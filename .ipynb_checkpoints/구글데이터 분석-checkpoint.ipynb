## 형변환2 : size를 int로 바꾼 뒤, 1m를 1로 잡고 변환
- df.ID = pd.to_numeric(df.ID, errors='coerce').fillna(0).astype(np.int64)
- print (df)

```python
size = df1['Size'].tolist()

for i,y in enumerate(size):
    if "k" in y:
        y = float(y.replace('k',''))/1024
        y = str("{0:.3f}M".format(y))
        size[i] = y
for i,y in enumerate(size):
    if "M" in y:
        y = float(y.replace('M',''))
        size[i] = y
    elif "Varies with device" in y:
        y = 0
        size[i] = y
for i, y in enumerate(size):
    if "1,000+" in  str(y):
        y = 1000
        size[i] = y
```

```python
df1['Size'] = size
```

## 형변환3 : 가격 타입 변환

```python
price = df1['Price'].tolist()

#가격 수정
for i, y in enumerate(price):
    if "$" in y:
        y = float(y.replace("$", ""))
        price[i] = y
    elif "Everyone" in y:
        y = 0
        price[i] = y
```

```python
# 아직은 '0'이  섞여있어 완벽한 float타입이 아닌 상황
# 1. price.replace('0', 0)   => 리스트는 replace가 안된다.
# 2. price = [w.replace('0', 0) for w in price] => for문으로 하나씩 뽑아내면 replace가 되긴 하지만 둘 다 스트링값으로 해야 한다.

# 일단 리스트를 행에다 대입
df1['Price'] = price
```

```python
# .replace로 바꾸는건 값 자체가 '0'일 때만 변환??   /    값 안에 있는 0을 변환하는건 .str.replace??
df1['Price'] = df1['Price'].replace('0', 0)
```
