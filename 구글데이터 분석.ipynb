## size를 int로 바꾼 뒤, 1m를 1로 잡고 변환
- df.ID = pd.to_numeric(df.ID, errors='coerce').fillna(0).astype(np.int64)
- print (df)

```python
size = df1['Size'].tolist()

for i,y in enumerate(size):
    if "k" in y:
        y = float(y.replace('k',''))/1024
        y = str("{0:.3f}M".format(y))
        size[i] = y
for i,y in enumerate(size):
    if "M" in y:
        y = float(y.replace('M',''))
        size[i] = y
    elif "Varies with device" in y:
        y = 0
        size[i] = y
for i, y in enumerate(size):
    if "1,000+" in  str(y):
        y = 1000
        size[i] = y
```

```python
df1['Size'] = size
```

## 가격 타입 변환

```python
price = df1['Price'].tolist()

#가격 수정
for i,y in enumerate(price):
    if "$" in y:
        y = float(y.replace("$",""))
        price[i] = y
    if '0' in y:
        y = 0
    elif "Everyone" in y:
        y = 0
        price[i] = y
```

```python
df1['Price'] = price
```
